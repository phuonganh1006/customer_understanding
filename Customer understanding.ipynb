#### Join & Expiry Date
expiry_date has null value (NaN).

If this member still active, expiry_date is null. 

```python
member[['customer_id', 'membership_status', 'join_date', 'expiry_date']]
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,5))
join_date = member.set_index('join_date')['customer_id'].resample('1Y').nunique()
expiry_date = member.set_index('expiry_date')['customer_id'].resample('1Y').nunique()
join_date.plot(kind='bar', ax=axs[0], rot=40)
expiry_date.plot(kind='bar', ax=axs[1], rot=40)
plt.show()
```

The two plots are very different, and both plots have their own meanings:

The picture on the left shows the lifespan of expired members. Most expired members only last for a short period of time

The picture on the right shows the lifespan of the entire member. In the entire distribution, there are 2 large clusters of customers (including expired and active members)

```python
fig, axs = plt.subplots(1,2, figsize=(10,5))
join_date = member['join_date']
expiry_date_with_null = member['expiry_date']
expiry_date_without_null = member['expiry_date'].fillna('2023-01-01')
result_with_null = ((expiry_date_with_null - join_date)/pd.Timedelta('1 days'))
result_without_null = ((expiry_date_without_null - join_date)/pd.Timedelta('1 days'))
sns.histplot(data=result_with_null, kde=True, ax=axs[0]).set(title='Member Lifespan (with null)')
sns.histplot(data=result_without_null, kde=True, ax=axs[1]).set(title='Member Lifespan (without null)')
plt.show()
```

```python
# Create new column member_lifespan
df['member_lifespan'] = result_without_null
```

### Churn-rate

In businesses, group members that exist for less than 3,6,12 months will create problems with customer retention.

In the picture on the left, for expired members, their lifespan only lasts for about 30-90 days.

```python
for g in member.groupby('membership_level'):
    num_total_member = len(g[1])
    num_expire = (g[1]['membership_status'] == 'Expired').sum() 
    rate_expire = round(num_expire/num_total_member * 100,2)
    print(f'Segment {g[0]} has total {num_total_member} customers and {num_expire} expired customers with general churn-rate {rate_expire}%')
    sns.histplot(data=g[1], x='member_lifespan', kde=True)
    plt.show()
```
